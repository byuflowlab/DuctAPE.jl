# Standard Expressions

```julia
### --- Geometry used everywhere --- ###
function relative_geometry(origin, field)
    # Extract Points
    xm = field[1]
    rm = field[2]
    
    xn = origin[1]
    rn = origin[2]

    # apply non-dimensionalization
    xi = (xm-xn)/rn
    rho = rm/rn
    
    # get least and greatest radial distance between field point and origin ring for convenience
    r1sq = xi^2 + (rho-1.0)^2
    r2sq = xi^2 + (rho+1.0)^2
    
    # get elliptic integral input
    k2 = 4.0*rho/r2sq
    
    # get elliptic integral values
    K = sf.ellipk(k2)
    E = sf.ellipe(k2)
    
    return (; xm, rm, xn, rn, xi, rho, r1sq, r2sq, k2, K, E)
end
```

```julia
### --- Vortex Rings --- ###
# - Unit Induced Axial Velocity - #
function vz_vortex(origin, field)
    
    # get the relative geometry
    (; xm, rm, xn, rn, xi, rho, r1sq, r2sq, k2, K, E) = relative_geometry(origin, field)
    
    if rm <= 0.0
        # avoid division by zero
        # TODO: figure out where this comes from (looks like standard expression for vortex filament)
        
        den = 2.0*rn(1.0+xi^2)^3/2
        
        vx = 1.0/den
        
    else
        # get pieces of expression to avoid mess
        den_out = 2.0*pi*rn*sqrt(r2sq)
        num_in = 2*(rho+1)
        den_in = r1sq

        # put things together
        vx = 1.0/den_out * (K - (1.0 + num_in/den_in)*E)
    end
    
    return vx
end


# - Unit Induced Radial Velocity - #
function vr_vortex(origin, field)
    
    # get the relative geometry
    (; xm, rm, xn, rn, xi, rho, r1sq, r2sq, k2, K, E) = relative_geometry(origin, field)
    
    if rm <= 0.0
        #avoid division by zero
        #no radial induction occurs at the axis
        
        vr = 0.0
    else
        # get pieces of expression to avoid mess
        num_out = xi
        den_out = 2.0*pi*rn*rho*sqrt(r2sq)
        num_in = 2*rho
        den_in = r1sq

        # put things together
        vr = -num_out/den_out * (K - (1.0 + num_in/den_in)*E)
    end
    
    return vr
end
```

```julia
# test that the functions run
origin = [0.0; 1.0]
field = [2.0; 2.0]
vx = vz_vortex(origin, field)
vr = vr_vortex(origin, field);
```

```julia
### --- look at point approached from sides --- ###
N = 100
fx = range(-1.0, 1.0; length=N)
origin = [0.0; 1.0]
vx = zeros(N)
vr = zeros(N)
for i in 1:N
    field = [fx[i]; origin[2]]
    vx[i] = vz_vortex(origin, field)
    vr[i] = vr_vortex(origin, field)
end

pvxh = plot(xlabel="x-distance", ylabel="unit axial induced velocity")
plot!(pvxh, fx,vx,color=blue[2],label="standard expression")

pvrh = plot(xlabel="r-distance", ylabel="unit radial induced velocity")
plot!(pvrh, fx,vr,color=blue[2],label="standard expression")

plot(pvxh,pvrh,layout=(2,1),size=(600,800))
```

```julia
### --- look at point approached from above/below --- ###
N = 100
fr = range(1.0, 3.0; length=N)
origin = [0.0; 2.0]
vx = zeros(N)
vr = zeros(N)
for i in 1:N
    field = [origin[1]; fr[i]]
    vx[i] = vz_vortex(origin, field)
    vr[i] = vr_vortex(origin, field)
end

pvxv = plot(xlabel="x-distance", ylabel="unit axial induced velocity")
plot!(pvxv, fr,vx,color=blue[2],label="standard expression")

pvrv = plot(xlabel="r-distance", ylabel="unit radial induced velocity")
plot!(pvrv, fr,vr,color=blue[2],label="standard expression")

plot(pvxv,pvrv,layout=(2,1),size=(600,800))
```

# Method of Matched Asymptotic Expansion

From "A panel method study of vortex sheets with special emphaisi on sheets of axisymmetric geometry" by Ichiro Sugioka and Sheila E. Widnall (NASA Contractor Report 177365)

More or less copied from their fortran code appendix for my purposes here.

```julia
function velin(r, x, y, dx, dy, w=1.0, g=-1.0)
    cost = dx/ sqrt(dx^2+dy^2)
    sint = dy/ sqrt(dx^2+dy^2)
    q5 = (x^2 + y^2)/w^2
    q4 = (y*cost - x*sint)/w
    q3 = (x*cost + y*sint)/w
    q2 = q5 - 2.0*q3 + 1.0
    q1 = q5 + 2.0*q3 + 1.0
    
    #check for singular condition
    if q4==0.0
        atn = 0.0
    else
        atn = atan((1.0-q3)/q4)+atan((1.0+q3)/q4)
    end
 
    # if q2 is zero, need to avoid log(0)
    if q2 == 0.0 || q1 == 0.0
        return 0.0, 0.0
    end
    
    #perturbed vortex strip formula
    u1 = -sint*log(q2/q1)+2.0*cost*atn
    v1 = -cost*log(q2/q1)-2.0*sint*atn
    u2 = (x/2.0/r)*u1
    v2 = (x/2.0/r)*v1 + 0.5*w/r * (log(q2*q1) - q3*log(q2/q1) + 2.0*q4*atn - 4.0)
    
    uin = -g/4.0/pi * (u1+u2)
    vin = g/4.0/pi * (v1+v2)
    
    return uin, vin
end

function velout(r, x, y, dx, dy, w=1.0, g=-1.0)
    r1 = sqrt(x^2 + y^2)
    r2 = sqrt((x+2.0*r)^2 + y^2)
    lam = (r2-r1)/(r1+r2) #TODO; WHAT IF LAM IS +/- 1, LEADING TO ELLIPTIC INTEGRALS TO BE INF??
    m1 = 1.0 - lam^2
    
    if r1==0.0 || lam==0.0
        return 0.0, 0.0        
    end
    
#     println("r1: ", r1)
#     println("r2: ", r2)
#     println("lam: ", lam)
#     println("m1: ", m1)
    
    dxr1 = x/r1
    dyr1 = y/r1
    dxr2 = (x+2.0*r)/r2
    dyr2 = y/r2
    dxl = (dxr2-dxr1)/(r2+r1) - lam/(r2+r1)*(dxr1+dxr2)
    dyl = (dyr2-dyr1)/(r2+r1) - lam/(r2+r1)*(dyr1+dyr2)
    K = sf.ellipk(lam^2)
    E = sf.ellipe(lam^2)
    Kp = E/lam/m1 - K/lam
    Ep = (E-K)/lam
    dxr = dxr1 + dxr2
    dyr = dyr1 + dyr2
    rr = r1 + r2
    
    uo = g*w/pi/(r+x)* (dyr*(K-E)+rr*(Kp-Ep)*dyl)
    vo = -g*w/pi/(r+x)* (dxr*(K-E)+rr*(Kp-Ep)*dxl)
    
    return uo, vo
end

function velio(r, x, y, dx, dy, w=1.0, g=-1.0)
    cost = dx/ sqrt(dx^2+dy^2)
    sint = dy/ sqrt(dx^2+dy^2)
    q5 = (r^2 + y^2)/w^2
    q4 = (y*cost - r*sint)/w
    q3 = (-r*cost + y*sint)/w
    q2 = q5 - 2.0*q3 + 1.0
    q1 = q5 + 2.0*q3 + 1.0
    
    r1 = sqrt(x^2 + y^2)
    if r1==0.0
        return 0.0, 0.0        
    end
   
    uio = -g*w/pi * (1.0+x/r/2.0)*y/r1/r1
    vio = g*w/pi * ((1.0 + x/r/2.0) * x/r1/r1 + 0.5/r*log(r1/8/r))
    vio += g*w/pi * log(8*r/w)/2.0/r
    
    return uio, vio
end

function v_comp(r, x, y, dx, dy, w=1.0, g=-1.0)
    vxi, vri = velin(r, x, y, dx, dy, w, g)
    vxo, vro = velout(r, x, y, dx, dy, w, g)
    vxio, vrio = velio(r, x, y, dx, dy, w, g)
    
    return vxi+vxo-vxio, vri+vro-vrio, vxi, vri, vxo, vro, vxio, vrio
end    
```

```julia
### --- look at horizontal panel approached from sides --- ###
N = 100
fx = range(-5.0, 5.0; length=N)
origin = [0.0; 1.0]
vx = zeros(N)
vr = zeros(N)
vxo = zeros(N)
vro = zeros(N)
vxi = zeros(N)
vri = zeros(N)
vxio = zeros(N)
vrio = zeros(N)
x = zeros(N)

r = origin[2]
y = 0.0
dx = 1.0
dy = 0.0
w = dx/2.0

for i in 1:N
    x[i] = fx[i]-origin[1]
    vx[i], vr[i], vxi[i], vri[i], vxo[i], vro[i], vxio[i], vrio[i] = v_comp(r, x[i], y, dx, dy, w)
end

println( v_comp(r, -2.0, y, dx, dy, w))
    
pvxc = plot(xlabel="x-distance", ylabel="unit axial induced velocity",grid=true)
plot!(pvxc, x,vxo,color=blue[1],label="outer expression")
plot!(pvxc, x,vxi,color=blue[2],label="inner expression")
plot!(pvxc, x,vxio,color=blue[3],label="matching expression")
plot!(pvxc, x,vx,color=red[2],label="combined expression")

pvrc = plot(xlabel="x-distance", ylabel="unit radial induced velocity",grid=true)
plot!(pvrc, x,vro,color=blue[1],label="outer expression")
plot!(pvrc, x,vri,color=blue[2],label="inner expression")
plot!(pvrc, x,vrio,color=blue[3],label="matching expression")
plot!(pvrc, x,vr,color=red[2],label="combined expression")

plot(pvxc,pvrc,layout=(2,1),size=(600,800))
```
